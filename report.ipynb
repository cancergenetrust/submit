# Steward Report
Crawl a steward generating reports on various things

```python
import re
import requests
import pandas as pd

# steward_multihash = "QmStJHN5CvcUbwkvVfMkCSmW4c9xgQRdMUv1r2h1VJxmZv"
steward_multihash = "QmTEp1XZG7A9qz15U1JjVBTJJdx1J9CNAWMaQ2aQhqqhTt"

# Multiple different IPFS gateways

# ipfs_url = "https://ipfs.infura.io/ipfs"
ipfs_url = "https://cloudflare-ipfs.com/ipfs"
# ipfs_url = "https://ipfs.io/ipfs"

timeout = 5*60
```

```python
index = requests.get("{}/{}".format(ipfs_url, steward_multihash)).json()
print("Domain {} has {} submissions".format(index['domain'], len(index['submissions'])))
```

```python
print("Submissions:")
submissions = []

for multihash in index["submissions"]:
    print(multihash)
    r = requests.get("{}/{}".format(ipfs_url, multihash), timeout=timeout)
    r.raise_for_status()
    submission = r.json()
    submission["multihash"] = multihash
    submissions.append(submission)
```

```python
pd.DataFrame([{"multihash": s["multihash"], "cgt_public_id": s["cgt_public_id"], "files": len(s["files"])}
              for s in submissions]).sort_values("cgt_public_id")
```

```python
# Load all foundation files
foundation_text = []
foundation = {}
for submission in submissions:
    for file in submission["files"]:
        if file["name"] == "foundation.json":
            print(file["multihash"])
            r = requests.get("{}/{}".format(ipfs_url, file["multihash"]), timeout=timeout)
            foundation_text.append(r.text)
            foundation[submission["multihash"]] = r.json()
```

```python
for f in foundation_text:
    if "TRF" in f:
        print(f)
```

```python
# Load all omop files
omop = {}
for submission in submissions:
    for file in submission["files"]:
        if file["name"] == "omop.json":
            print(submission["multihash"])
            r = requests.get("{}/{}".format(ipfs_url, file["multihash"]), timeout=timeout)
            r.raise_for_status()
            omop[submission["multihash"]] = r.json()
```

```python
def walk_dictionary(input):
    for key,value in input.items():
        if isinstance(value, dict):
            walk_dictionary(value)
        elif isinstance(value, list):
            for i in value:
                walk_dictionary(i)
        else:
            if isinstance(value, str):
                if "http" in value:
                    continue
                if len(value) > 37:
                    print("long string", key, len(value), value)
                if re.findall(r"\/.*?\/", value):
                    print("possible date", value, len(value))
                if re.findall(r"\-.*?\-", value):
                    print("possible date", value, len(value))
                if re.findall(r"age|mrn|trf", value, re.IGNORECASE):
                    print("other", value, len(value))
                    
                    
for k, v in omop.items():
    print(">>> Submission:", k)
    walk_dictionary(v)

# for k, v in foundation.items():
#     print(">>> Submission:", k)
#     walk_dictionary(v)
```
